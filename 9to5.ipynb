1. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Write a query that returns the series_id, industry_code, industry_name, and value from the january_2017 table but only if that value is greater than the average value for annual_2016 of data_type_code 82.</span>

```sql
select distinct j.series_id,i.industry_code,industry_name,j.value
from LaborStatisticsDB.dbo.january_2017 j,LaborStatisticsDB.dbo.industry i,LaborStatisticsDB.dbo.series s
where j.series_id=s.series_id AND 
      i.industry_code=s.industry_code and 
            j.value> (
                      select avg(a.value)
                      from LaborStatisticsDB.dbo.annual_2016 a,LaborStatisticsDB.dbo.series s ,LaborStatisticsDB.dbo.datatype d
                    where d.data_type_code ='82'
                 and s.series_id =a.series_id 
                 and d.data_type_code=s.data_type_code)
                 

```

2. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Create a Union table comparing average weekly earnings&nbsp; of production and nonsupervisory employees between annual_16 and january_17 using the data type 30.&nbsp; Round to the nearest penny.&nbsp; You should have a column for the average earnings and a column for the year, and the period.</span>

```sql
select round(avg(value),2) Avg,period, year
from LaborStatisticsDB.dbo.annual_2016 a,LaborStatisticsDB.dbo.series s,LaborStatisticsDB.dbo.datatype d
where a.series_id=s.series_id AND 
     s.data_type_code=d.data_type_code and 
    d.data_type_code='30' 
    and lower(series_title) ='average weekly earnings of production and nonsupervisory employees'
group by period ,year



UNION
select round(avg(value),2) avg,period, year
from LaborStatisticsDB.dbo.january_2017 j,LaborStatisticsDB.dbo.series s1,LaborStatisticsDB.dbo.datatype d1
where j.series_id=s1.series_id AND 
    s1.data_type_code=d1.data_type_code and 
     d1.data_type_code='30' 
   and lower(series_title) ='average weekly earnings of production and nonsupervisory employees'
group by period,year

      
```
